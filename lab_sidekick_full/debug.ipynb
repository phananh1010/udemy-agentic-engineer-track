```python
### Test core of sidekick
```

```python
sidekick = Sidekick()
await sidekick.setup()
print (sidekick.sidekick_id, sidekick.graph)
print ("Success if have valid uuid and CompliedStateGraph object")
```

```python
### Perform basic operation
```

```python
message = "hello?"
success_criteria = "be polite and concise"
history = []

config = {"configurable": {"thread_id": sidekick.sidekick_id}}
state = {"messages": message,
         "success_criteria": success_criteria or "The answer should be clear and accurate",
         "feedback_on_work": None,
         "success_criteria_met": False,
         "user_input_needed": False,
        }
result = await sidekick.graph.ainvoke(state, config=config)
print (result.keys())
print ("Success if have all keys like this: dict_keys(['messages', 'success_criteria', 'feedback_on_work', 'success_criteria_met', 'user_input_needed'])")
```
